```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 
import xgboost
from sklearn.ensemble import RandomForestRegressor
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pandas as pd
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import Imputer
from sklearn.model_selection import cross_val_score

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
print(os.listdir("../input"))
dataTrain = pd.read_csv("../input/train.csv")
dataTest = pd.read_csv("../input/test.csv")
print(dataTrain.columns)
columnsOfInterest = ['Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare', 'Embarked']
trainY = dataTrain.Survived

trainX = dataTrain[columnsOfInterest]
oheTrainX = pd.get_dummies(trainX)



oheTestX = pd.get_dummies(dataTest[columnsOfInterest])

my_imputer = Imputer()
oheTrainX = my_imputer.fit_transform(oheTrainX)

myModel = XGBClassifier(n_estimators = 1000, learning_rate = 0.01)
myModel.fit(oheTrainX, trainY, verbose = False)
preds = myModel.predict(oheTrainX)
#preds = np.int_(np.around(preds, decimals = 0))
model = xgboost.XGBClassifier()
#model = RandomForestRegressor()



scores = cross_val_score(model, oheTrainX, trainY, scoring='neg_mean_absolute_error')
print(scores)
print('Mean Absolute Error %2f' %(-1 * scores.mean()))





#print(mean_absolute_error(preds, testY))
test = pd.read_csv('../input/test.csv')
testX = test[columnsOfInterest]
ohetestX = pd.get_dummies(testX)
ohetestX = my_imputer.fit_transform(ohetestX)
#print(ohetestX.head())
predSurvived = myModel.predict(ohetestX)
predSurvived = np.int_(np.around(predSurvived, decimals = 0))
print(predSurvived)
my_submission = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predSurvived})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)

# Any results you write to the current directory are saved as output.
```
